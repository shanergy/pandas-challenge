```python
# calculate "Per Student Budget" and place into a new column
schools_df["budget_per_student"] = (schools_df["budget"] / schools_df["size"]).astype(float)
schools_df.head()
```

```python
# add two new columns: "passing_reading_flag" and "passing_reading_flag"
students_df["passing_reading_flag"] = [1 if x >= 70 else 0 for x in students_df["reading_score"]]
students_df["passing_math_flag"] = [1 if x >= 70 else 0 for x in students_df["math_score"]]
students_df.head()
```

```python
# combine data into one dataset
combinedRaw_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# calculate the total number of schools
totalSchools = df["school_name"].nunique()
# totalSchools = df["School ID"].nunique()
# totalSchools = len(df["school_name"].unique())
# totalSchools = len(df["School ID"].unique())
# totalSchools
```

```python
# calculate the total number of students
totalStudents = df["Student ID"].nunique()
# totalStudents = df["Student ID"].count()
# totalStudents
```

```python
# calculate the total budget
totalBudget = schools_df["budget"].sum()
# totalBudget
```

```python
# calculate the average math score
avgMathScore = df["math_score"].mean()
# avgMathScore
```

```python
# calculate the average reading score
avgReadingScore = df["reading_score"].mean()
# avgReadingScore
```

```python
# calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
avgOverallPassingRate = (df["math_score"].mean() + df["reading_score"].mean()) / 2
# avgOverallPassingRate
```

```python
# calculate the percentage of students with a passing math score (70 or greater)
passingMath = (df["math_score"] >= 70).value_counts()[True]
# passingMath
pctPassingMath = passingMath / totalStudents * 100
# pctPassingMath
```

```python
# calculate the percentage of students with a passing reading score (70 or greater)
passingReading = (df["reading_score"] >= 70).value_counts()[True]
# passingReading
pctPassingReading = passingReading / totalStudents * 100
# pctPassingReading
```

```python
# create a summary dataframe to hold the summary results for the district
districtSummary_df = pd.DataFrame({"Total Schools": [totalSchools],
                                   "Total Students": [totalStudents],
                                   "Total Budget": [totalBudget],
                                   "Per Student Budget": (totalBudget / totalStudents),
                                   "Average Math Score": [avgMathScore],
                                   "Average Reading Score": [avgReadingScore],
                                   "% Passing Math": [pctPassingMath],
                                   "% Passing Reading": [pctPassingReading],
                                   "% Overall Passing Rate": [avgOverallPassingRate]
                                  })

districtSummary_df["Total Students"] = districtSummary_df["Total Students"].astype(int).map("{:,}".format)
districtSummary_df["Total Budget"] = districtSummary_df["Total Budget"].astype(float).map("${:,.2f}".format)
districtSummary_df["Per Student Budget"] = districtSummary_df["Per Student Budget"].astype(float).map("${:,.2f}".format)

# districtSummary_df
# districtSummary_df.transpose()  # <--- this is another veiw; interesting to see it transposed
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# use .groupby to separate data into fields according to "school_name" values
grouped_schools_df = df.groupby(["school_name"])
# grouped_schools_df = df.groupby(["School ID", "school_name", "type", "size", "budget"])
# grouped_schools_df

# to visualized, must use a function:
# grouped_schools_df.count()
```

```python
# get the school type for each school
schoolsType = grouped_schools_df["type"].unique()

# calculate the total number of students per school
schoolsStudentCounts = grouped_schools_df["Student ID"].nunique()

# get the budget for each school (using average/.mean() as there is one distinct value for each school)
schoolsTotalBudget = grouped_schools_df["budget"].mean()

# get the budget_per_student for each school (using average/.mean() as there is one distinct value for each school)
schoolsBudgetPerStudent = grouped_schools_df["budget_per_student"].mean()

# calculate the average math score per school
schoolsAvgMathScores = grouped_schools_df["math_score"].mean()

# calculate the average reading score per school
schoolsAvgReadingScores = grouped_schools_df["reading_score"].mean()

# calculate the number of students with a passing math score per school
schoolsPassingMath = grouped_schools_df["passing_math_flag"].sum()

# calculate the number of students with a passing reading score per school
schoolsPassingReading = grouped_schools_df["passing_reading_flag"].sum()
```

```python
# create a summary dataframe to hold the summary results for each school
schoolsSummary_df = pd.DataFrame({"School Type": schoolsType,
                                  "Total Students": schoolsStudentCounts,
                                  "Total School Budget": schoolsTotalBudget,
                                  "Per Student Budget": schoolsBudgetPerStudent,
                                  "Average Math Score": schoolsAvgMathScores,
                                  "Average Reading Score": schoolsAvgReadingScores,
                                  "% Passing Math": schoolsPassingMath / schoolsStudentCounts * 100,
                                  "% Passing Reading": schoolsPassingReading / schoolsStudentCounts * 100,
                                  "% Overall Passing Rate": (schoolsPassingMath + schoolsPassingReading) / schoolsStudentCounts / 2 * 100
                                  })

schoolsSummary_df["Total Students"] = schoolsSummary_df["Total Students"].astype(int).map("{:,}".format)
schoolsSummary_df["Total School Budget"] = schoolsSummary_df["Total School Budget"].astype(float).map("${:,.2f}".format)
schoolsSummary_df["Per Student Budget"] = schoolsSummary_df["Per Student Budget"].astype(float).map("${:,.2f}".format)

# delete the name of the index (for cleaner look and to match example)
del schoolsSummary_df.index.name
# schoolsSummary_df
```
